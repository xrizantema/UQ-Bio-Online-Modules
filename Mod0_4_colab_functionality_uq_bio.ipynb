## Magic functions

There are also a handful of magic functions denoted by a ```%``` or ```%%``` in interactive Python notebooks. For example ``%timeit`` lets you time functions without importing time and using that yourself. timeit will run the codeblock or cell 10000's of times and report back an average time to run that function. The double ```%%``` denotes that this command runs in a cell wide basis, for example the ```%%``` version of timeit, ```%%timeit```, keeps track of how long the whole cell takes to run, not just the function behind it. The single ```%``` is called a "line magic" and a double ```%%``` is called a "cell magic." 

`````python
import numpy as np

# run the magic function time it to time this line of code!
%timeit np.random.randint(0,100,size=(1,10000))

# run the magic function time it to time this line of code!
%timeit np.random.randint(0,100,size=(1,1000))
`````

`````python
%%timeit 
# %% double % denotes this function is run cell wide, so this function times this entire cell
np.random.randint(0,100,size=(1,10000))
np.random.randint(0,100,size=(1,1000))
np.random.randint(0,100,size=(1,100))
`````

## tqdm integration

```tqdm``` is a python package that gives access to a nice progress bar for running large computations. It can be used to give feedback in a google colab (as well as a local python kernel).

`````python
import tqdm.notebook as tq #import tqdm notebook version
import time
def waiting_function():
  time.sleep(1)

for i in tq.tqdm(range(0,15)):
  waiting_function()

for i in tq.trange(0,15):
  waiting_function()
`````

`````python
#@title numpy ndarray of shape T, X, Y, C --> MP4 video
from matplotlib import animation
from IPython.display import HTML

def make_movie(data, total_frames,  blit=True, interval=100 ):
  
  fig,axes = plt.subplots(1,3,dpi=120,tight_layout=True,)
  #define initial frame here 
  #ex: axes.plot(data[0],data[1])

  def frame_generator(frame_number):
    ## Define plot of image as function of frame number here

  anim = animation.FuncAnimation(fig, frame_generator, frames=total_frames, interval=interval, blit=blit)
  return HTML(anim.to_html5_video())




def make_movie_image_tensor(image_tensor, cmaps = ['Reds_r','Greens_r','Blues_r'], channels = True):
  '''
  makes an html5 video of an image tensor of shape:
        [t, x, y, color]

  channels = True to plot channels separately 
  '''
  if channels:  # if channels = True we will make a 3 color / 3 plot movie
    fig,axes = plt.subplots(1,3,dpi=120,tight_layout=True,)
  else: # otherwise just one subplot is needed
    fig,axes = plt.subplots(1,1,dpi=120,tight_layout=True,) #tight layout sets the axis's apart
  if channels: #if channels set up initial images here
    im1 = axes[0].imshow(image_tensor[0,:,:,0], cmap=cmaps[0])  #get blank axes.image objects to maniuplate later
    im2 = axes[1].imshow(image_tensor[0,:,:,1], cmap=cmaps[1])
    im3 = axes[2].imshow(image_tensor[0,:,:,2], cmap=cmaps[2])
    for k in range(3):
      axes[k].set_title('Channel %i'%k)
  else:
    im1 = axes.imshow(image_tensor[0,:,:,:], cmap=cmaps[0]) #if merge just make one
    axes.set_title('Merge')
  plt.close()

  #here set up the function that is called by FuncAnimation, it returns the artist objects and updates the frame by i
  def movieFrame(i):
    ## Extract all the image channels
    #Define  frames
    if channels:  #get the channels if 3 color
      channel_1_img = image_tensor[i,:,:,0]
      channel_2_img = image_tensor[i,:,:,1]
      channel_3_img = image_tensor[i,:,:,2]
      images = [channel_1_img, channel_2_img, channel_3_img]  # concatenate channels
      axes_handles = [im1,im2,im3]  #Axes handles from the imshow
    else:
      channel_img = image_tensor[i,:,:,:]   
      images = [channel_img,]  # concatenate channels
      axes_handles = [im1,]  #Axes handles from the imshow

    #update the axes imshow objects with the new data per k channels
    for k,image_n in enumerate(images):  #for loop that returns an object and an index
      axes_handles[k].set_array(images[k])  # 
    return axes_handles  #return the object
  
  #here call the animation function from matplotlib and then return the final HTML video object to the user
  anim = animation.FuncAnimation(fig, movieFrame, frames=image_tensor.shape[0], interval=100, blit=True)
  return HTML(anim.to_html5_video())

`````
